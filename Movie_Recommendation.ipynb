```python
#generes,id,keywords,title,overview,release_date,cast,crew

movie_df=movie_df[['genres','id','keywords','title','overview','cast','crew']]
```

# Data Preprocessing Steps

```python
display(movie_df.isnull().sum())
display(movie_df.duplicated().sum())
```

```python
movie_df=movie_df.dropna(subset=['overview'])
```

```python
movie_df.iloc[0].genres
```

```python
from typing import List
import ast
def convert_by_name(generes_data)->List[str]:
  return [item['name'] for item in ast.literal_eval(generes_data)]
```

```python
movie_df['genres']=movie_df['genres'].apply(convert_by_name)
```

```python
movie_df['keywords']=movie_df['keywords'].apply(convert_by_name)
```

```python
movie_df['cast']=movie_df['cast'].apply(lambda x: convert_by_name(x)[:3])
```

```python
def find_director(crew_data):
  return [item['name'] for item in ast.literal_eval(crew_data) if item['job']=='Director']
```

```python
movie_df['crew']=movie_df['crew'].apply(find_director)
```

```python
movie_df['overview']=movie_df['overview'].apply(lambda x:x.split())
```

```python
movie_df
```

```python
movie_df['tags']=movie_df['overview']+movie_df['genres']+movie_df['keywords']+movie_df['cast']+movie_df['crew']
```

```python
movie_new_df=movie_df[['id','title','tags']]
```

```python
movie_new_df.loc[:, 'tags'] = movie_new_df['tags'].apply(
    lambda x: " ".join(x).lower()
)
```

```python
movie_new_df.head()
```

# Using Vectorization for similartity words

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(
    max_features=5000,
    stop_words='english'
)

tfidf_matrix = tfidf.fit_transform(movie_new_df['tags'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(tfidf_matrix)
```

```python
def recommend_movies(movie_title, df, similarity_matrix, top_n=5):
    if movie_title not in df['title'].values:
        return "Movie not found"

    idx = df[df['title'] == movie_title].index[0]

    similarity_scores = list(enumerate(similarity_matrix[idx]))
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)

    similarity_scores = similarity_scores[1: top_n + 1]

    movie_indices = [i[0] for i in similarity_scores]

    return df.iloc[movie_indices][['title']]
```

```python
title_name=movie_new_df['title'].values.tolist()
title_name
```

```python
display(recommend_movies(
    movie_title="Pirates of the Caribbean: At World's End",
    df=movie_new_df,
    similarity_matrix=cosine_sim,
    top_n=5
))
```
